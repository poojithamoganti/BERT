**Converting sentences to tokens and tokens to id's**

```python
from transformers import BertModel, AutoTokenizer
```

```python
import pandas as pd
```

```python
model_name = "bert-base-cased"
```

Bert Model

```python
model = BertModel.from_pretrained(model_name)
```

```python
print(model)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
print(tokenizer)
```

```python
sentence = "When life gives you lemons, make lemonade"
```

Breaking down sentence to tokens

```python
tokens = tokenizer.tokenize(sentence)
```

```python
print(tokens)
```

For every word there will be a vocablory, as each word has a particular id assigned to it.

```python
vocab = tokenizer.vocab
```

```python
print(vocab)
```

```python
vocab_df = pd.DataFrame({"token":vocab.keys(), "token_id":vocab.values()})
```

```python
vocab_df = vocab_df.sort_values(by='token_id').set_index('token_id')
```

```python
print(vocab_df)
```

```python
token_id = tokenizer.encode(sentence)
```

Token ids with respect to the tokens in the sentence


```python
print(token_id)
```

```python
len(tokens)
```

```python
len(token_id)
```

```python
vocab_df.iloc[101]
```

```python
vocab_df.iloc[102]
```

List of tokens with their id's

```python
list(zip(tokens, token_id[1:-1]))
```

Decoding back to words/sentence

```python
tokenizer.decode(token_id[1:-1])
```

Adding words back to sentence



```python
sentence2 = sentence.replace("make","make it a")
```

```python
print(sentence2)
```

Input ids, token ids, attention masks

```python
tokenizer_out2 = tokenizer([sentence, sentence2], padding = True)
print(tokenizer_out2)
```

To view Padding tokens and converting id's back to sentences


```python
tokenizer.decode(tokenizer_out2['input_ids'][1])
```
